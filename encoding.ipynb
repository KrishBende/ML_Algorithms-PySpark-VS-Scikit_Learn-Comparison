```python
import pandas as pd
```

```python
original_file="data/CollegePlacement.csv"
```

```python
df=pd.read_csv(original_file)
print(df.isna().sum())
print(df.shape)
print(df.head(5))
#df = df.replace('?', None)     #For adult.csv
df.dropna(inplace=True)
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder

print("Original DataFrame:")
print(df.head(2))

#Label Encoding
object_cols = df.select_dtypes(include='object').columns
le = LabelEncoder()
for col in object_cols:
    df[col] = le.fit_transform(df[col])

print("\nDataFrame after Label Encoding:")
print(df.head(2))
```

```python
df.to_csv("data/encoded_college_placement.csv",index=False)
```

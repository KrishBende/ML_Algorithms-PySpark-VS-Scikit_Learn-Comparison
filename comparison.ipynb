```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Income Dataset

models=['Logistic Regression','Decision Tree','SVM','Naive Bayes','Random Forest']
sklearn_accuracies=[79.19,81.22,79.59,79.47,85.42]
mllib_accuracies=[82.25,84.07,81.06,79.07,83.88]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - width/2, sklearn_accuracies, width, label='SKLearn')
rects2 = ax.bar(x + width/2, mllib_accuracies, width, label='MLlib')

# Add labels, title, and other customizations
ax.set_ylabel('Accuracy')
ax.set_title('Annual Income Dataset')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
